```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import chromedriver_autoinstaller
```

```python
chromedriver_autoinstaller.install()

browser = webdriver.Chrome()
```

```python
browser.get('https://www.olx.co.id/jakarta-selatan_g4000030/motor-bekas_c200')
```

```python
soup = BeautifulSoup(browser.page_source, 'lxml')
```

```python
rows = soup.find_all('div', class_='fTZT3')
```

```python
len(rows)
```

```python
data = []
for row in rows:
    harga = row.find(class_='_2Ks63').text
    tahun = row.find(class_='YBbhy').text
    jenis_brand = row.find(class_='_2poNJ').text
    lokasi = row.find(class_='_2VQu4').text
    
    dt = {
        'harga': harga,
        'tahun': tahun,
        'jenis_brand': jenis_brand,
        'lokasi': lokasi
    }
    
    data.append(dt)   
```

```python
df1 = pd.DataFrame(data)
```
